```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/preprocessed.csv", index_col=0)
data.head()
```

```python
cols = data.dtypes[data.dtypes == "object"].index
for col in cols:
    data[col] = data[col].apply(lambda x: x.strip())
data.head()
```

### Relation between Rating and MRP

```python
sns.lineplot(x=data.Rating, y=data.MRP)
plt.show()
```

### Price Variation with Processor

```python
chart = sns.barplot(x=data.Processor, y=data.MRP)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

### Price Variation with OS

```python
chart = sns.barplot(x=data.OS, y=data.MRP)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

### Price Variation due to Brand Influence

```python
chart = sns.barplot(x=data.Brand, y=data.MRP)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

### Ram Size influence on Price

```python
chart = sns.barplot(x=data.RAM, y=data.MRP)
plt.show()
```
